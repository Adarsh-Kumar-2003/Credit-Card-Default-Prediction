```python

#function for graph generation of precision recalls and thresholds
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
 plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
 plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
 plt.xlabel("Threshold")
 plt.legend(loc="upper left")
 plt.ylim([0, 1])
```

**LOGISTIC REGRESSION MODEL**

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
predictions_test = []
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
fold = 1

for train_idx, test_idx in skf.split(X_train_pca, y_train):
    print("fold:- ", fold)
    X_s_train, X_s_test = X_train_pca[train_idx], X_train_pca[test_idx]
    y_s_train, y_s_test = y_train.iloc[train_idx], y_train.iloc[test_idx]


    model_logistic = LogisticRegression(max_iter=1000,class_weight='balanced')
    model_logistic.fit(X_s_train, y_s_train)
    y_predict_probab = model_logistic.predict_proba(X_s_test)
    predictions_test.append(y_predict_probab)
    y_probs = model_logistic.predict_proba(X_s_test)[:, 1]
    y_predictions_set = model_logistic.predict(X_s_test)


    auc = roc_auc_score(y_s_test, y_probs)

    print("roc score:- ", auc)
    print("precision:- ", precision_score(y_s_test,y_predictions_set))
    print("recall:- ", recall_score(y_s_test,y_predictions_set))
    print("f1_score:- ", f1_score(y_s_test,y_predictions_set))
    y_scores = model_logistic.decision_function(X_s_test)
    precisions, recalls, thresholds = precision_recall_curve(y_s_test, y_scores)
    plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
    plt.show()
    fold += 1
```

**XGB Classifier Model**

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

fold = 1
for train_idx, test_idx in skf.split(X_train_pca, y_train):
    print("fold:-", fold)

    X_s_train, X_s_test = X_train_pca[train_idx], X_train_pca[test_idx]
    y_s_train, y_s_test = y_train.iloc[train_idx], y_train.iloc[test_idx]
    model_xgb = XGBClassifier(
        n_estimators=100,
        max_depth=6,
        learning_rate=0.1,
        scale_pos_weight=scale_pos_weight,
        use_label_encoder=False,
        eval_metric='logloss',
        n_jobs=-1,
        random_state=42
    )

    model_xgb.fit(X_s_train, y_s_train)


    y_probs = model_xgb.predict_proba(X_s_test)[:, 1]
    y_preds = model_xgb.predict(X_s_test)
    roc = roc_auc_score(y_s_test, y_probs)
    precision = precision_score(y_s_test, y_preds, zero_division=0)
    recall = recall_score(y_s_test, y_preds)
    f1 = f1_score(y_s_test, y_preds)
    precisions, recalls, thresholds = precision_recall_curve(y_s_test, y_probs)
    plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
    plt.show()

    print("ROC AUC Score:-", roc)
    print("Precision:-", precision)
    print("Recall:- ", recall)
    print("F1 Score:-", f1)

    fold += 1
```
